```python
import warnings
warnings.filterwarnings('ignore')

import plotly.express as px
import pandas as pd
import numpy as np
df=px.data.gapminder()
```

```python
is_asia=df['continent']=='Asia'
df=df[is_asia]
is_year=df['year']>1993
df=df[is_year]
Adf=df[['country','year','gdpPercap']]
```

```python
pdf = pd.pivot_table(Adf, index='country', columns='year', values='gdpPercap')
```

```python
pdf.reset_index()
```

```python
fig = px.scatter_ternary(pdf, a="1997", b="2002", c="2007", color="year", 
                        hover_name="country",
                         color_discrete_map={"1997":"blue", "2002":"green", "2007":"red"})
fig.show()
```

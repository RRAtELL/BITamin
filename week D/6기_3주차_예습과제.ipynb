```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
iris_data=iris.data
iris_df=pd.DataFrame(data=iris_data, columns=iris.feature_names)
print('<데이터의 평균 값>')
print(iris_df.mean())
print('\n<데이터의 분산 값>')
print(iris_df.var())
iris_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(iris_df)
iris_scaled=scaler.transform(iris_df)
iris_df_scaled=pd.DataFrame(data=iris_scaled,columns=iris.feature_names)
print('<데이터의 평균 값>')
print(iris_df_scaled.mean())
print('\n<데이터의 분산 값>')
print(iris_df_scaled.var())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(iris_df)
iris_scaled=scaler.transform(iris_df)
iris_df_scaled=pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print('<정규화된 데이터의 최댓값>')
print(iris_df_scaled.max())
print('\n<정규화된 데이터의 최솟값>')
print(iris_df_scaled.min())
```

```python
iris_df_abs_scaled=iris_df/abs(iris_df.max())
print('<abs 정규화된 데이터의 최댓값>')
print(iris_df_abs_scaled.max())
print('\n<abs 정규화된 데이터의 최솟값>')
print(iris_df_abs_scaled.min())
```

```python
an_array=np.random.rand(10)*10
print('<원 배열>\n, an_array')
norm=np.linalg.norm(an_array)
print('<원 배열의 Norm> \n', norm)
normal_array=an_array/norm
print('<벡터정규화된 배열>\n', normal_array)
print('<벡터정규화된 배열의 Norm>', np.linalg.norm(normal_array))
```

```python
iris_df_vector_scaled=iris_df/np.linalg.norm(iris_df)
print("<벡터 정규화된 데이터의 최댓값>")
print(iris_df_vector_scaled.max())
print("<벡터 정규화된 데이터의 최솟값>")
print("<원 데이터 프레임의 Norm>", np.linalg.norm(iris_df))
print("<벡터 정규화된 데이터 프레임의 Norm>", np.linalg.norm(iris_df_vector_scaled))
```

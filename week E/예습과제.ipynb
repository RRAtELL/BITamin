```python
import pandas as pd
import seaborn as sns
titanic=sns.load_dataset('titanic')
titanic=titanic.loc[:,['age','sex','class','fare','survived']]
titanic.head()
```

```python
pdf1=pd.pivot_table(titanic, index='class', columns='sex', values='age', aggfunc='mean')
print(pdf1.head())
```

```python
pdf2=pd.pivot_table(titanic, index='class', columns='sex', values='survived', aggfunc=['mean', 'sum'])
pdf2.head()
```

```python
pdf3=pd.pivot_table(titanic, index=['class','sex'], columns=['survived'], values=['age','fare'], aggfunc=['mean','max'])
pdf3.head()
```

```python
pdf3.xs('First')
pdf3.xs(('First','female'))
```

```python
pdf3.xs('male', level='sex')
pdf3.xs(('Second','male'), level=['class','sex'])
pdf3.xs(('Second','male'), level=[0,1])
```

```python
pdf3.xs('mean', axis=1)
pdf3.xs(('mean', 'age'),axis=1)
```

```python
pdf3.xs(1, level='survived', axis=1)
pdf3.xs(1, level=2, axis=1)
pdf3.xs(('max', 'fare', 0), level=[0,1,2], axis=1)
```
